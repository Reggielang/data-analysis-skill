```python
import pandas as pd
import numpy as np
position = pd.read_csv('position.csv',encoding='utf-8')
company = pd.read_csv('company_utf.csv',encoding='utf-8')
```

```python
#建立数据透视表
position.pivot_table(index='city',columns='workYear',values='avg')
```

```python
position.pivot_table(index=['city','education'],
                     columns='workYear',
                     values=['avg','top'],
                     aggfunc = [np.mean,np.sum],margins=True)
```

```python
#对值的针对性计算
position.pivot_table(index=['city','education'],
                     columns='workYear',
                     values=['avg','top'],
                     aggfunc = {'avg':np.mean,'top':np.sum},margins=True)
```
