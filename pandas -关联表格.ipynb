```python
import pandas as pd
position = pd.read_csv('position.csv',encoding='utf-8')
company = pd.read_csv('company_utf.csv',encoding='utf-8')
```

```python
position.head()
```

```python
company.head()
```

```python
position.merge(right=company,how ='inner',on= 'companyId')
```

```python
#修改列名比较简单的方法
col = list(company.columns)
col[0]= 'id'
company.columns = col
```

```python
company.head()
```

```python
#merge合并表格针对某个列
position.merge(right=company,how ='inner',left_on= 'companyId',right_on = 'id')
```

```python
#join合并表格是针对索引
company.join(position)
```

```python
#concat类似于表格的堆叠,axis为0是上下堆叠，1是左右堆叠。
pd.concat([company,position],axis = 1)
```
