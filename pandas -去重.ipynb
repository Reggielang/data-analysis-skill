```python
import pandas as pd
import numpy as np
position = pd.read_csv('position.csv',encoding='utf-8')
company = pd.read_csv('company_utf.csv',encoding='utf-8')
```

```python
#人为造成一点缺失值
position.loc[position.city == '深圳','city'] = np.NaN
```

```python
#填充，或者删除缺失值所在列或者行,axis =1, 会删除列
#fillna()
position.dropna()
```

```python
#找出重复值
position[position.duplicated()]
```

```python
s =pd.Series([1,1,2,3,4])
```

```python
#~代表反向取值，这里代表取出了不重复的值
s[~s.duplicated()]
```

```python
#代表表示留下最后一次的重复值
s[~s.duplicated(keep='last')]
```

```python
#直接把重复值删除了。
s.drop_duplicates()
```
