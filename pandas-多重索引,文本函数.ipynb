```python
import pandas as pd
position = pd.read_csv('position.csv',encoding='utf-8')
company = pd.read_csv('company_utf.csv',encoding='utf-8')
```

```python
#series形式的多重索引
position.groupby(by = ['city','education']).mean().avg['上海']
```

```python
#dataframe形式的多重索引
position.groupby(by = ['city','education']).mean().loc['上海']
```

```python
position.sort_values(by=['city','education']).set_index(['city','education'])
```

```python
position.groupby(by = ['city','education']).mean().reset_index()
```
