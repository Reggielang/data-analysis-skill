```python
import pandas as pd
df = pd.read_csv('position.csv',encoding='utf-8')
```

```python
%matplotlib inline
```

# 各种图的代码
    -折线图plot
    -柱状图bar
    -直方图hist
    -箱线图box
    -密度图kde
    -面积图area
    -散点图scatter
    -散点图矩阵 scatter_matrix
    -饼图 pie

```python
#计数之后，还要进行排序，因为索引现在是无序的
df.avg.value_counts().sort_index().plot()
```

```python
df.avg.value_counts().sort_index().plot.bar()
```

```python
df.pivot_table(index='city',
               columns='education',
               values='avg',
               aggfunc='count').plot.bar(stacked = True)
```

```python
df.groupby('education').apply(lambda x :x.avg).unstack().T.plot.hist(alpha = 0.8,stacked= True,bins = 30)
```

```python
df.boxplot(column='avg',by = 'education')
```

```python
df.avg.plot.kde()
```

```python
df.pivot_table(index = 'avg',columns='education',values='positionId',aggfunc='count').plot.area()
```

```python
df.groupby('companyId').agg(['mean','count','max']).avg.plot.scatter(x= 'mean',y='count')
matrix = df.groupby('companyId').agg(['mean','count','max']).avg
```

```python
#散点图矩阵，如果有多个值，想要了解值之间的关系，使用矩阵比较合适。
pd.plotting.scatter_matrix(matrix.query('count<50'),diagonal='kde')
```

```python
df.city.value_counts().plot.pie(figsize= (6,6))
```
