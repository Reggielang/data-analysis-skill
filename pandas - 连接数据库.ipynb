```python
#使用pandas连接数据库
import pandas as pd
from sqlalchemy import create_engine
```

```python
sql = 'select * from company'
engine = create_engine('mysql+pymysql://root:123456@localhost:3306/data?charset=utf8')
df = pd.read_sql(sql, engine)
```

```python
df
```

```python
def reader(query,db):
    sql = query
    engine = create_engine('mysql+pymysql://root:123456@localhost:3306/{0}?charset=utf8'.format(db))
    df = pd.read_sql(sql, engine)
    return df
```

```python
reader('select * from company','data')
```

```python
df_company = reader('select * from company','data')
df_dataanalyst = reader('select * from dataanalyst','data')
```

```python
#直接合并两个表格会报错，因为两个df中companyId数据值类型有差异,所以同时转换为int，之后再进行合并
df_company['companyId'] = df_company['companyId'].apply(int)
df_dataanalyst['companyId'] = df_dataanalyst['companyId'].apply(int)
merged  = pd.merge(df_dataanalyst,df_company, on = 'companyId')
merged
```

```python
#m每个公司的不同职位的计数
result = merged.groupby(['city','companyFullName']).count()['positionId'].reset_index()
```

```python
result.info()
```

```python
#把结果写入到数据库中
result.to_sql(name = 'new_table',con = 'mysql+pymysql://root:123456@localhost:3306/data?charset=utf8',
               if_exists='append',index=False)
```
