```python
import pandas as pd
df = pd.read_csv('DataAnalyst_gbk.csv',encoding='gbk',sep= ',')
df
```

```python
df.groupby(by = 'city')
```

```python
df.avg.value_counts()
```

```python
df.groupby(by='city')['avg'].max()
```

```python
df.groupby(['city','workYear']).max()
```

```python
#根据groupby之后得出的是tuple，然后v仍然是dateframe，所以找出评价薪资之后，使用max和min函数，相减即可求出差值。
for k,v in df.groupby(by='city'):
    print(max(v.avg),min(v.avg))
    print(max(v.avg)-min(v.avg))
    print('-------------------')
```
