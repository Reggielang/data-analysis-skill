# 图表构成
-title

-data

-x轴
    x轴刻度 xtick
    X轴标签 xlabel
    
-y轴
    y轴刻度 ytick
    y轴标签 ylabel

```python
#一个图表最基础的构成属性
plt.figure(1,figsize=(10,4))
plt.plot(np.random.random_integers(-10,10,20))
plt.title('折线图')
plt.xticks([0,10,20])
plt.xlabel('x轴')
plt.show()
```

```python
plt.plot(np.random.random_integers(-10,10,20),label = 'no1')
plt.plot(np.random.random_integers(-10,10,20),label = 'no2')
plt.legend(('no1','no2'))
plt.show()
```

```python
data = df.groupby(['education','companyId']).agg(['mean','count']).avg.reset_index()
```

```python
#根据不同的学历，把他们进行聚合为不同的数据框，然后分别求出他们的平均值，和数量，然后再作图。
for edu,group in data.groupby('education'):
    x = group['mean']
    y = group['count']
    plt.scatter(x,y,label = edu)
plt.xlabel('平均工资')
plt.ylabel('招聘数')
plt.legend()
plt.show()    
```

```python
#一张画布里，显示多个图表，subplot中2代表2行，2代表俩列，1代表左边的图。
plt.figure(figsize=(12,4))
#第一张图
plt.subplot(2,2,1)
plt.plot(np.random.random_integers(-10,10,20),label = 'no1')
plt.plot(np.random.random_integers(-10,10,20),label = 'no2')
#第二个图
plt.subplot(2,2,2)
plt.plot(np.random.random_integers(-10,10,20),label = 'no3')
plt.plot(np.random.random_integers(-10,10,20),label = 'no4')
plt.legend()

#第三个大图
plt.subplot(2,1,2)
plt.plot(np.random.random_integers(-10,10,20),label = 'no5')

plt.show()
```

```python
#一张画布里，显示多个图表，subplot中1代表1行，2代表俩列，1代表左边的图。
plt.figure(figsize=(12,4))
#第一张图
plt.subplot(1,2,1)
plt.plot(np.random.random_integers(-10,10,20),label = 'no1')
plt.plot(np.random.random_integers(-10,10,20),label = 'no2')
#第二个图
plt.subplot(1,2,2)
plt.plot(np.random.random_integers(-10,10,20),label = 'no3')
plt.plot(np.random.random_integers(-10,10,20),label = 'no4')
plt.legend()

plt.show()
```

```python
#一张画布里，显示多个图表，subplot中2代表2行，2代表俩列，1代表左边的图。
plt.figure(figsize=(12,4))
#第一张图
plt.subplot(2,2,1)
plt.plot(np.random.random_integers(-10,10,20),label = 'no1')
plt.plot(np.random.random_integers(-10,10,20),label = 'no2')
#第二个图，第二个图占据了2x2的单元格中的第三个位置
plt.subplot(2,2,3)
plt.plot(np.random.random_integers(-10,10,20),label = 'no3')
plt.plot(np.random.random_integers(-10,10,20),label = 'no4')
plt.legend()

#第三个大图
plt.subplot(1,2,2)
plt.plot(np.random.random_integers(-10,10,20),label = 'no5')

plt.show()
```
