```python
import pandas as pd
import numpy as np
position = pd.read_csv('position.csv',encoding='utf-8')
company = pd.read_csv('company_utf.csv',encoding='utf-8')
```

```python
position.avg.apply(lambda x : str(x)+'k')
```

```python
def func(x):
    if x > 20:
        return '20+k'
    else:
        return '0~20k'
position.apply(lambda x :func(x.avg),axis=1)    
```

```python
#定义一个查找薪资前五的函数
def func(x):
    r = x.sort_values('avg',ascending = False)
    return r[:5]
```

```python
position.groupby('city').apply(func)
```

```python
#前五的薪资排序
position.sort_values('avg',ascending = False)[:5]
```

```python
def func(x,n):
    r = x.sort_values('avg',ascending = False)
    return r[:n]
```

```python
position.groupby('city').apply(func,n = 3)
```

```python
position.groupby('city').agg(['sum','mean'])
```
